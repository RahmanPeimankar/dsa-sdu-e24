```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
        else:
            current = self.head
            while current.next != None:
                current = current.next
            current.next = new_node

    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def delete(self, data):
        if self.head == None:
            return

        if self.head.data == data:
            self.head = self.head.next
            return

        current = self.head
        while current.next != None:
            if current.next.data == data:
                current.next = current.next.next
                return
            current = current.next

    def printout(self):
        current = self.head
        while current != None:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# As discussed in the lecture
# find second to last by checking if next.next is none

def find_second_to_last(linked_list):
        # No particular need to have these be two different if statements when an "or" works
        # There will not be a second to last if either the head or head.next is none.
        # That would mean that the linked list is empty or only contains one element respectively.
        if linked_list.head == None or linked_list.head.next == None:
            return None

        # Checking to see if next.next is none.
        # When it is, the second to last element in the linked list has been found.
        current = linked_list.head
        while current.next.next != None:
            current = current.next

        return current


def concatenate_linked_lists(headL, headM):
    # Making the linked list L' (LPrime) 
    listLP = LinkedList()
    # Go through the list L starting at the head and append each element to list LP
    current = headL
    while current != None:
        listLP.append(current.data)
        current = current.next
    # Same deal with list M
    current = headM
    while current != None:
        listLP.append(current.data)
        current = current.next

    return listLP

# Testcode starts here

# Create and fill list L
listL = LinkedList()
for i in range (9):
    listL.append(i)


second_to_last_element = find_second_to_last(listL)

if second_to_last_element != None:
    print("The second-to-last element in list L is:", second_to_last_element.data)
else:
    print("The stack is too short to have a second-to-last element.")


# Create and fill list M
listM = LinkedList()
for i in range (6):
    listM.append(i+5)

# concatenating list L and list M and printing the result
listLP = concatenate_linked_lists(listL.head, listM.head)
print("Concatenated List:", end=" ")
listLP.printout()
```
